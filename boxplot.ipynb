```python
# Dependencies
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
file=Path("Resources/City_council_housing_data_2018_2021.csv")
house=pd.read_csv(file,index_col=False)
house
```

```python
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'With train' column to numerical values: Yes=1, No=0
house['With train'] = house['With train'].map({'Yes': 1, 'No': 0})

# Now compute the correlation
correlation_coefficient, p_value = pearsonr(house['With train'], house['Total Number of Sales'])

print("Correlation Coefficient:", correlation_coefficient)
print("P-value:", p_value)

# Visualizing with a boxplot
sns.boxplot(x='With train', y='Total Number of Sales', data=house)
plt.show()
```
